Рассчитаем скользащее среднее для окна равного 6 часам. Чем интервал больше, тем сильнее сглаживание. Возможно, для дальнейшей работы будет выгодно оценить скользящее среднее для окна, равного 24 часам.

Для более удобной визуализации построим скользящее среднее для окна в 24 часа.

```python
eda_data['rolling_mean'] = eda_data['num_orders'].rolling(24).mean()
eda_data['std'] = eda_data['num_orders'].rolling(24).std()
```

```python
eda_data.plot(figsize=(10,5), grid=True)
plt.title('Временной ряд [num_orders],\n'
          'скользящее среднее [rolling_mean]'
          ' с окном 6 часов\n'
          'и скользящее стандартное отклонение[std]')
plt.xlabel('Дата и время')
plt.ylabel('Количество заказов')
plt.show()
```

Видим некоторые аномально большие значения количества заказов в некоторые периоды. Уточним, в какое время было столько заказов. При этом указанные выбросы на скользящее среднее сильного влияния не оказывают.

```python
eda_data[eda_data['num_orders'] > 300]
```

Вероятно в августе было какое-либо знаменательное событие, что три недели подряд было очень много заказов в одно время. Отметим этот факт и при обучении моделей проверим влияние на результаты обучения.

Изучим тренды и сезонность заказов такси

```python
decomposed = seasonal_decompose(eda_data['num_orders'])
plt.figure(figsize=(8.5, 8))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca(), grid=True)
plt.title('Тренд временного ряда')
plt.xlabel('Дата и время')
plt.ylabel('Количество заказов')
plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca(), grid=True)
plt.title('Сезонность временного ряда')
plt.xlabel('Дата и время')
plt.ylabel('Количество заказов')
plt.subplot(313)
decomposed.resid.plot(ax=plt.gca(), grid=True)
plt.title('Остатки')
plt.xlabel('Дата и время')
plt.ylabel('Количество заказов')
plt.tight_layout()
```

Получившийся **тренд** очень похож на скользящее среднее с окном в 24 часа. Считаем, что такое окно наиболее подходящее для дальнейшей работы.

**Сезонность** по кварталам, месяцам, неделям не выявлена. Сезонность целесообразно оценить по **дням** за неделю.

```python
decomposed.seasonal['2018-03-01':'2018-03-06'].plot(figsize=(10, 5), grid=True)
plt.title('Сезонность временного ряда\nдля окна в одни сутки в мае')
plt.xlabel('Дата и время')
plt.ylabel('Количество заказов')
plt.show()
```

В то же время, оценим среднее количество заказов по часам, дням недели и дням месяца. 

```python
eda_data.groupby(eda_data.index.hour).mean().rename(
    columns={'num_orders':'Количество заказов'}
)['Количество заказов'].plot(
    figsize=(10, 5),
    title='Распределение заказов по часам в течение суток', 
    xlabel='Часы суток', 
    ylabel='Количество заказов', 
    xlim=(0, 23), 
    grid=True
);
```

Каждый день мы можем наблюдать одинаковую картину. Около 00:00 каждый день среднее количество заказов максимально и резко снижается к 06:00, затем неравномерно растет с провалами около 12:00, 18:00.

```python
eda_data.groupby(eda_data.index.weekday).mean().rename(
    columns={'num_orders':'Количество заказов'}, 
    index={0:'Понедельник', 1:'Вторник', 2:'Среда', 3:'Четверг', 4:'Пятница', 5:'Суббота', 6:'Воскресенье'}
)['Количество заказов'].plot(
    figsize=(10, 5),
    title='Распределение заказов по дням в течение недели', 
    xlabel='Дни недели', 
    ylabel='Количество заказов', 
    xlim=(0, 6), 
    grid=True
);
```

Видно, что среднее количество заказов такси высоко в понедельник, во вторник заказов минимальное количество, затем показатель расчет до максимума в пятницу (который очень близок к значению в понедельник) и снижается в выходные до уровня вторника.

```python
eda_data.groupby(eda_data.index.day).mean().rename(
    columns={'num_orders':'Количество заказов'}
)['Количество заказов'].plot(
    figsize=(10, 5),
    title='Распределение заказов по дням в течение месяца', 
    xlabel='Числа месяца', 
    ylabel='Количество заказов', 
    xlim=(1, 31), 
    grid=True
);
```

Видно, что среднее количество заказов в начале месяца минимально и затем растет к концу месяца.

Сформируем разницу между количеством заказов и уровнем тренда (сколзящее среднее за неделю).

```python
eda_data['num_orders_monthly'] = eda_data['num_orders'] - eda_data['num_orders'].rolling(168).mean()
```

```python
eda_data.groupby(eda_data.index.day).mean().rename(
    columns={'num_orders_monthly':'Количество заказов'}
)['Количество заказов'].plot(
    figsize=(10, 5),
    title='Распределение заказов по дням в течение месяца', 
    xlabel='Числа месяца', 
    ylabel='Количество заказов', 
    xlim=(1, 31), 
    grid=True
);
```

Устойчивый паттерн выявить не удается. Действительно, подъем к концу месяца связан именно с общим трендом, который мы исключили. Целесообразно в признаках не учитывать день месяца.

Предварительно создадим копию входных данных для текущего этапа работы с учетом того, что на этапе EDA мы не вносили изменения.

```python
process_data = preprocessing_data.copy()
process_data.head()
```

Создадим новые признаки:
1. В связи с тем, что временной диапазон составляет несколько месяцев 2018 года, для признаков целесообразно использовать показатели месяца, дня и часа.
2. Дополнительным признаком примем день недели.
3. Добавим "отстающие значения". Количество отстающих значений целесообразно выбрать равным окну скользящего среднего.
4. В рамках исследовательного анализа данных скользящее среднее уже было включено в дата-сет, как признак, задающий общий тренд временного ряда. Учтем, что вычисление скользящего среднего не должно включать в себя текущее значение ряда.

Создадим функцию для формирования признаков.

```python
def make_features(data, max_lag, rolling_mean_size):
    data = data.copy()
    # Исключим месяц
    # data['month'] = data.index.month
    # Исключим признак день, поскольку зависимости 
    # от дня месяца не обнаружено
    # data['day'] = data.index.day
    data['hour'] = data.index.hour
    data['dayofweek'] = data.index.dayofweek 
        
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    num_columns = []
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
        num_columns.append('lag_{}'.format(lag))
    
    for lag in range(48, 24*7+1, 24):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
        num_columns.append('lag_{}'.format(lag))
        
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    return data, num_columns
```

Устойчивый паттерн выявить не удается. Действительно, подъем к концу месяца связан именно с общим трендом, который мы исключили. Целесообразно в признаках не учитывать день месяца.

Сформируем итоговый дата-сет

```python
final_data, num_columns = make_features(process_data, 24, 24)
final_data.head()
```

```python
num_columns.append('rolling_mean')
```

Удалим пустые значения

```python
final_data = final_data.dropna()
final_data.head()
```

```python
final_data.shape
```

```python
def load_pipe(num_scaler,
              num_columns,
              model=None):

    data_preprocessor = ColumnTransformer(
        [
            ('num', num_scaler, num_columns)
        ], 
        remainder='passthrough'
    )
    
    pipe = Pipeline(
        [
            ('preprocessor', data_preprocessor),
            ('models', model)
        ]
    )
    
    return pipe
```

**Однако, тестовую выборку мы не должны применять сейчас. Лучше провести кросс-валидацию.**

Поэтому с учетом следующего замечания попробуем применить TimeSeriesSplit. А чтобы запустить кросс-валидацию для моего "медианного оценщика" - создадим его.

```python
class MedianEstimator(BaseEstimator):

    def __init__(self, constant=None):
        self.constant = constant
        
    def fit(self, X, y):

        # Check that X and y have correct shape
        X, y = check_X_y(X, y)

        self.X_ = X
        self.y_ = y
        if self.constant is None:
            self.constant = np.median(y)
            
        # Return the classifier
        return self

    def predict(self, X):

        # Check if fit has been called
        check_is_fitted(self)

        # Input validation
        X = check_array(X)

        predictions = np.ones(X.shape[0]) * self.constant
        return predictions
```

```python
model_median = MedianEstimator()
```

```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
cv_result_median = cross_val_score(
    model_median, 
    X_train,
    y_train, 
    cv=tscv,  
    scoring='neg_root_mean_squared_error', 
    verbose=500
)
```

```python
print(
    'Среднее значение RMSE: '
    f'{(-1)*cv_result_median.mean():.2f}'
     )
print(
    'Стандартное отклонение предсказаний: '
    f'{cv_result_median.std():.2f}'
)
```

Таким образом, мы получили результаты кросс-валидации временного ряда (с использованием TimeSeriesSplit) для созданного оценщика.

```python
results = [
    (
        'Constant',
        None,
        None,
        round((-1)*cv_result_median.mean(), 3),
        round(cv_result_median.std(), 3)
    )
]
```
